```python
#read csv
import pandas as pd
import re
from bs4 import BeautifulSoup
import nltk
from nltk.corpus import stopwords
train = pd.read_csv("labeledTrainData.tsv",header=0,delimiter="\t",quoting=3)

#get train data file
train.columns.values

#read review colums
# train["review"][0]

#remove html tag
reviewExample = BeautifulSoup(train["review"][0]) #set to html format
reviewExample.get_text() #get html text without all html tag

#remove all none english words .all punctuation
#use regex expression
letterOnly = re.sub("[^a-zA-Z]"," ",reviewExample.get_text()) #all not letter reply with " "

#make all letter lowercase 
txt_lower = letterOnly.lower()
words = txt_lower.split() # make string split with " "

#remove all stop word: a the etc using NL Toolkit
#download the data set form ntkt
nltk.download("stopwords")
```

```python
train.shape
```

```python
#remove all stop in all split words list
words = [w for w in words if not w in stopwords.words('english')]
print(words)
```

```python
#define a function above
def reviewToWords(raw_reviews):
  #to remove html tag and get the txt
  reviews = BeautifulSoup(raw_reviews)
  
  #remove all punctuation
  letterOnly = re.sub("[^a-zA-z]", " ",reviews.get_text())

  #splite the words to list
  words = letterOnly.lower().split()

  #clean the stop word
  stops = set(stopwords.words("english"))
  words = [w for w in words if not w in stops]

  #return the string 
  return (" ".join(words))
```

```python
#get total length and loop over all data
total_size = train["review"].size

all_review_data = []


for i in range(total_size):
  all_review_data.append(reviewToWords(train["review"][i]))
  if i%1000 == 0:
    print("%d of %d"%(i+1,total_size))
```

```python
#now we need to make those word to 'Bag of word
#meaning:each word appear time in a sentance
#check the word to tokenization
from sklearn.feature_extraction.text import CountVectorizer
#tokenization : max 5000 words
vectorizer = CountVectorizer(analyzer="word",tokenizer=None,preprocessor = None,stop_words = None,max_features=5000)

#and now we fit the transform model and it will learn and chang the word to token vector
train_feature = vectorizer.fit_transform(all_review_data)
train_feature = train_feature.toarray() #convert to a list
```

```python
train_feature[0] #first sentance vector
```

```python
train_feature.shape
```

```python
vectorizer.get_feature_names() #show all words
```

```python
#all sentance is changed to vector,we need to fit the model with this value
#using Random Forest Classifier (a tree struct)
from sklearn.ensemble import RandomForestClassifier

#init RFC
forest = RandomForestClassifier(n_estimators=100) #with 100 trees(default 100)

#we need to fit this model with out sentance vector and lable is train set [sentiment]
forest = forest.fit(train_feature,train["sentiment"])
```

```python
# now we need to use our test set to test the model
# load the test data set
test = pd.read_csv("testData.tsv", header=0, delimiter="\t", quoting=3)
test_length = len(test["review"])
all_test_review = []
for i in range(test_length):
  all_test_review.append(reviewToWords(test["review"][i]))

#now we have all clean work
#and change to vector
#we now in test set ,no need to fit the else will case overfitting
test_feature = vectorizer.transform(all_test_review)
test_feature = test_feature.toarray()

#now we need to predict using our trained model(RFC)
#predict
result = forest.predict(test_feature)

#output the reuslt to csv
output = pd.DataFrame(data={"id":test["id"],"sentiment":result})
output.to_csv("Bag_of_words.csv",index=False,quoting=3)

```
